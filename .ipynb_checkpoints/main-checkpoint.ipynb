```python
spider_types = ("black widow", "funnel web", "brown recluse",
                "bold jumper", "european garden")
```

```python
path = Path("spiders")
```

```python
if not path.exists():
    path.mkdir()
    for spider in spider_types:
        dest = (path/spider)
        dest.mkdir(exist_ok=True)
        results = search_images_bing(key, spider + " spider")
        download_images(dest, urls = results.attrgot("content_url"))
        
```

```python
filenames = get_image_files(path)
filenames
```

```python
failed = verify_images(filenames)
failed
```

```python
failed.map(Path.unlink)
```

```python
spiders = DataBlock(blocks = (ImageBlock, CategoryBlock),
                   get_items = get_image_files,
                   splitter = RandomSplitter(valid_pct=0.2, seed=0),
                   get_y = parent_label,
                   item_tfms = Resize(224))
```

```python
dls = spiders.dataloaders(path)
```

```python
dls.valid.show_batch(max_n=4, nrows=1)
```

```python
learn = cnn_learner(dls, resnet18, metrics = error_rate)
```

```python
learn.fine_tune(4)
```

```python
interp = ClassificationInterpretation.from_learner(learn)
cm = interp.confusion_matrix()
```

```python
cm
```

```python
correct = np.diagonal(cm)
```

```python
print("Total : {}".format(cm.sum()))
print("Correct Prediction : {}".format(correct.sum()))
print("False Prediction : {}".format(cm.sum() - correct.sum()))
print("Accuracy : {:.2f}".format(correct.sum()/cm.sum() ))
```

```python
interp.plot_top_losses(5, nrows=5)
```
